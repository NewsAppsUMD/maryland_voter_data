```python
import pandas as pd
df = pd.read_csv("md_voter_file.txt", sep='\t', on_bad_lines='skip') 
df.head()
```

```python
df.columns
```

```python
from datetime import date
from datetime import datetime
# coded based on https://www.geeksforgeeks.org/python-program-to-calculate-age-in-year/

def calculateAge(birthDate):
    today = date.today()
    age = today.year - birthDate.year - ((today.month, today.day) <(birthDate.month, birthDate.day))
    return age
     
```

```python
import warnings
warnings.filterwarnings('ignore')
df["BirthDate"] = pd.to_datetime(df["BirthDate"], format="%m/%d/%Y")
df['age'] = df['BirthDate'].apply(calculateAge)
```

```python
df['age_bracket'] = df['age'].apply(lambda x: 
    "18-24" if x <= 24 else 
    "25-34" if x <= 34 else 
    "35-50" if x <= 50 else 
    "51-64" if x <= 64 else 
    "65+")
```

```python
df.head()
```

```python
df.columns
replacements = {key: 'OTH' for key in [
    'ONLM', 'OIN', 'OWCP', 'OLB', 'GRN',
    'OBAR', 'OSC', 'OCH', 'OPAT', 'ORE',
    'OCON', 'OSD', 'OFCH', 'OLIR', 'OCN',
    'OCS', 'OAM', 'OTP', 'OCO', 'OAL',
    'OAI', 'OPOP', 'ORL', 'OWO', 'OBM',
    'ONL', '21', 'OID', 'OAN', 'OPE',
    'OWH', 'ONP', 'OFR', 'ONAS'
]}

# Replace values in the specific column
df['Party'] = df['Party'].replace(replacements)
```

```python
df.columns
df2 = df.groupby(["age_bracket", "Gender","County", "Party"]).agg(Count=("FirstName", "count"), voted_2024 = ("11/05/2024-PG","count"),voted_2020 = ("11/03/2020-PG","count"),voted_2016 = ("11/08/2016-PG","count")).reset_index()

df3 = df.groupby(["age_bracket","County", "Party"]).agg(Count=("FirstName", "count"), voted_2024 = ("11/05/2024-PG","count"),voted_2020 = ("11/03/2020-PG","count"),voted_2016 = ("11/08/2016-PG","count")).reset_index()

df3.head()
```

```python
df3.to_csv('agg_MD_noGenderSplit_County.csv')
```

```python
import pandas as pd
df2 = pd.read_csv("agg_MD_noGenderSplit_County.csv")
df2
```

```python
df3 = df2.groupby(["County", "age_bracket"]).agg(vote_2024=("voted_2024", "sum")).reset_index()
df3
```

```python
grouped_df = df3.groupby(['County', 'age_bracket'], as_index=False)['vote_2024'].sum()
grouped_df['total_votes'] = grouped_df.groupby('County')['vote_2024'].transform('sum')
grouped_df['turnout'] = grouped_df['vote_2024'] / grouped_df['total_votes']
grouped_df
pivoted_df = grouped_df.pivot(index='County', columns='age_bracket', values='turnout').reset_index()

# Display the pivoted DataFrame
print(pivoted_df)
```
