```python
import pandas as pd
df = pd.read_csv("calvert_voters.csv")
df
```

```python
from datetime import date
from datetime import datetime
# coded based on https://www.geeksforgeeks.org/python-program-to-calculate-age-in-year/

def calculateAge(birthDate):
    birthDate = datetime.strptime(birthDate, "%Y-%m-%d").date()
    today = date.today()
    age = today.year - birthDate.year - ((today.month, today.day) <(birthDate.month, birthDate.day))
    return age
     
print(calculateAge("2021-02-03"))
```

```python
import warnings
warnings.filterwarnings('ignore')
df['age'] = df['birth_date'].apply(calculateAge)
```

```python
df['age_bracket'] = None
for i in range(len(df)):
    df['age_bracket'][i] = "18-24" if df["age"][i] <= 24 else "25-34" if df["age"][i] <= 34 else "35-50" if df["age"][i] <= 50 else "51-64" if df["age"][i] <= 64 else "65+"
```
